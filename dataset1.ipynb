```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # Imported Just In Case I Need It
import seaborn as sns # Imported Just In Case I Need It

if __name__=="__main__":
    # Reading the D1-Questionnaire dataset.
    questionnaire_result = pd.read_csv("./datasets/D1-Questionnaire.csv")
    
    # Cleaning the D1-Questionnaire dataset by removing all rows that have N/A values.
    cleaned_data=questionnaire_result.dropna()
    
    # Sorting the dawta by age and removing the 'participant' column
    cleaned_data=cleaned_data.sort_values(by=['age'])
    cleaned_data=cleaned_data.drop(columns=['participant'])
    # Creating Gender Dataframes based on the conditions of 'gender' column
    male_participants=cleaned_data[(cleaned_data['gender']=='m')]

    print("MALE PARTICIPANTS")
    print(male_participants.head())
    
    female_participants=cleaned_data[(cleaned_data['gender']=='f')]
    
    print("FEMALE PARTICIPANTS")
    print(female_participants.head())
    
    # Creating Education Dataframes based on the conditions of 'education' column
    bachelor_participants=cleaned_data[(cleaned_data['education']=='Bachelor')]
    
    print("BACHELOR PARTICIPANTS")
    print(bachelor_participants.head())
    
    master_participants=cleaned_data[(cleaned_data['education']=='Master')]
    
    print("MASTER PARTICIPANTS")
    print(master_participants.head())
    
    abitur_participants=cleaned_data[(cleaned_data['education']=='Abitur')]
    
    print("ABITUR PARTICIPANTS")
    print(abitur_participants.head())
    
    staatsexamen_participants=cleaned_data[(cleaned_data['education']=='Abitur/Staatsexamen')]
    
    print("ABITUR/STAATSEXAMEN PARTICIPANTS")
    print(staatsexamen_participants.head())
```
